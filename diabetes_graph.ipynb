```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12
```

```python
df = pd.read_csv('diabetes.csv')

print("Veri setinin ilk 5 satırı:")
df.head()
```

```python
outcome_counts = df['Outcome'].value_counts()

# Sütun grafiği
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=['Diyabet Değil', 'Diyabet'], y=outcome_counts.values, palette=['#5ab4ac', '#d8b365'])

for i, v in enumerate(outcome_counts.values):
    ax.text(i, v + 5, str(v), ha='center', fontweight='bold')

plt.title('Diyabet Olan ve Olmayan Hasta Sayısı', fontsize=16, fontweight='bold')
plt.xlabel('Durum', fontsize=12)
plt.ylabel('Hasta Sayısı', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
df['Age_Group'] = pd.cut(df['Age'], bins=[20, 30, 40, 50, 60, 70, 80, 90], 
                         labels=['21-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90'])

age_diabetes = df.groupby(['Age_Group', 'Outcome']).size().unstack()

plt.figure(figsize=(12, 7))
age_diabetes.plot(kind='bar', stacked=False, ax=plt.gca(), color=['#5ab4ac', '#d8b365'])

plt.title('Yaş Gruplarına Göre Diyabet Dağılımı', fontsize=16, fontweight='bold')
plt.xlabel('Yaş Grubu', fontsize=12)
plt.ylabel('Hasta Sayısı', fontsize=12)
plt.legend(['Diyabet Değil', 'Diyabet'])

plt.tight_layout()
plt.show()
```

```python
df['Glucose_Group'] = pd.cut(df['Glucose'], bins=[0, 50, 100, 150, 200], 
                            labels=['0-50', '51-100', '101-150', '151-200'])

glucose_diabetes = df.groupby(['Glucose_Group', 'Outcome']).size().unstack()

plt.figure(figsize=(12, 7))
glucose_diabetes.plot(kind='bar', stacked=False, ax=plt.gca(), color=['#5ab4ac', '#d8b365'])

plt.title('Glikoz Seviyesine Göre Diyabet Dağılımı', fontsize=16, fontweight='bold')
plt.xlabel('Glikoz Seviyesi', fontsize=12)
plt.ylabel('Hasta Sayısı', fontsize=12)
plt.legend(['Diyabet Değil', 'Diyabet'])

plt.tight_layout()
plt.show()
```
